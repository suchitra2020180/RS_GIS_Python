```python
def download(url):
  filename=os.path.join(data_folder_path,os.path.basename(url))
  with requests.get(url,stream=True, allow_redirects=True) as r:
    with open(filename,'wb') as file:
      for chunks in r.iter_content(chunk_size=8096):
        file.write(chunks)
      print(f'Downloaded successfully {filename}')
```

Here the Dadara & Nagar Haveli is saved in 2 lines so replace Dadara & Nagar Haveli with Dadara & Nagar Haveli and Dman & Diu (UT) and remove Nan

```python
main_df=df.copy()
main_df['State/Union Territory']=main_df['State/Union Territory'].apply(lambda row:'Dadara & Nagar Haveli and Daman & Diu (UT)' if row=='Dadara & Nagar Haveli' else row )
main_df['State/Union Territory']=main_df['State/Union Territory'].apply(lambda row:'West Bengal' if row=='Wes t Bengal' else row )
#main_df['State/Union Territory']=main_df['State/Union Territory'].str.replace(' ','')
#main_df['State/Union Territory']=main_df['State/Union Territory'].str.replace(r' (?!(\&))|(?<!(&)) ', '', regex=True)

# Function to conditionally remove spaces
def remove_spaces_conditional(statename):
    if '&' in statename or ' and ' in statename:
        # If '&' or 'and' is present, return the string as it is
        return statename
    else:
        # Remove all spaces
        return statename.replace(' ', '')

# Apply function to 'statename' column
main_df['State/Union Territory'] = main_df['State/Union Territory'].apply(remove_spaces_conditional)
main_df['State/Union Territory']=main_df['State/Union Territory'].apply(lambda row:'Jammu & Kashmir (UT)' if row=='Jam mu & Kashmir (UT)' else row )


#Now remove Nan values
main_df.dropna(inplace=True)
main_df.shape
```

##Converting address to coordinates

```python
#Using open route services
ORS_API_Key = 'ORS_API_Key'


def geocode(location):
  url = 'https://api.openrouteservice.org/geocode/search'
  parameters={
      'api_key':ORS_API_Key,
      'text':location
  }
  response=requests.get(url,parameters)
  if response.status_code==200:
    data=response.json()
    if data['features']:
      coordinates=data['features'][0]['geometry']['coordinates']
      return coordinates #returns [longitude,latitude]

```
