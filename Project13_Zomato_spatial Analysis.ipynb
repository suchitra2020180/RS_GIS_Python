```python
#Lets extract lat,long from address column
#Create a new empty dataframe
loc_coordinates=pd.DataFrame()
loc_coordinates['Name']=df['location'].unique()
loc_coordinates
```

```python
#To convert lation to lat,long we will use geopy package
```

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
geolocator=Nominatim(user_agent='app',timeout=None)
```

```python
#Lets create 2 empty list to store lat, long values
lat=[]
long=[]

for name in loc_coordinates['Name']:
    location=geolocator.geocode(name)  #geographcal format from open street Map
    #print('location:',location)
    #print('Location latitude',location.latitude)
    
    if location is None:
        lat.append(np.nan)
        long.append(np.nan)
    else:
        lat.append(location.latitude)
        long.append(location.longitude)
        
loc_coordinates['Latitude']=lat
loc_coordinates['Longitude']=long
loc_coordinates
```
